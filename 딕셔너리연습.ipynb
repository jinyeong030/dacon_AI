딕셔너리 데이터 추가


```python
fruits = {"사과":1000}
print("기존 과일 가격표:", fruits)
#add ftuit
fruits["바나나"]= 2000
fruits["블루베리"]=6000
fruits["딸기"]=13000

print("업데이트 과일 가격표:", fruits)
```

딕셔너리 데이터 수정 (단일)

```python
fruits["사과"] = 1300
print("가격인상 과일 가격표:", fruits)
#update()
fruits.update({
    "사과":700,
    "바나나":1500,
    "블루베리":5500,
    "딸기":10000
})
print("전체 세일 과일 가격표:", fruits)
```

딕셔너리 데이터 삭제

```python
phone={
    "준혁":"010-1234-1234",
    "민석":"010-9876-9876",
    "서준":"010-1111-2222",
    "남사친":"010-2222-3333",
    "남친":"010-0000-0000"
}
print("삭제전: ",phone)

#del사용
del phone["남친"]
print("삭제후: ",phone)

#pop() 사용
del_num = phone.pop("남사친")
print("삭제한 번호: ",del_num)
print("현재 번호: ",phone)

#clear() 딕셔너리 데이터 전체 삭제
phone.clear()
print("삭제후: ",phone)
```

keys() 딕셔너리의 모든 키 가져오기
values()
items()

```python
fruits = {"사과":1000,"딸기":12000,"바나나":2500,"블루베리":7000}

fruit_names = fruits.keys()
print("과일 목록: ",fruit_names)

for fruit in fruits.keys():
    print(f"과일 이름: {fruit}")

prices = fruits.values()
print("과일 가격:",prices)
print("과일 총 가격:",sum(prices))

for name,price in fruits.items():
    print(f"{name}가격: {price}원")
    
```

```python
lunch = {
    "월요일":"김치찌개",
    "화요일":"제육볶음",
    "수요일":"돈까스"
}
#1 메뉴 확인
print("===오늘의 메뉴===")
today = "월요일"
print(f"{today} 메뉴는 {lunch[today]}입니다.")

#2 메뉴 추가
print("\n===메뉴 추가===")
lunch["목요일"]="김밥"
print("메뉴가 추가되었습니다.")
print(lunch)

#3메뉴 변경
print("\n===메뉴 변경===")
lunch["화요일"]="비빔밥"
print("메뉴가 변경되었습니다.")
print(lunch)

#특정 요일 메뉴 확인
print("\n===수요일 메뉴 확인===")
print(f"수요일 메뉴는 {lunch["수요일"]}입니다.")
```
