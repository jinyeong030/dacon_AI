딕셔너리 반복

```python
student = {"이름":"철수","나이":12,"학년":6}

#key값
for key in student.keys():
    print(key)

#value값
for value in student.values():
    print(value)

#key & value
for key,value in student.items():
    print(f"{key}: {value}")
```

✋ break 이해하기


```python
numbers = [1, 2, 3, 4, 5]

for number in numbers:
    if number == 3:
        break
    print(number)
```

⏭ continue 이해하기

```python
numbers = [1, 2, 3, 4, 5]

for number in numbers:
    if number == 3:
        continue
    print(number)
```

실전 예제: 스마트 에어컨 시스템 시뮬레이션


```python
import random

threshold_temp = 30        # 온도가 30도 이상이면 '높다'고 판단

print("=== 실내 온도 모니터링을 시작합니다! ===")

while True:
    current_temp = random.randint(15,33)
    print(f"현재 온도: {current_temp}도")

    if current_temp >= threshold_temp:
        print(f"온도가 {current_temp}도로 30도 이상입니다.")
        print("에어컨을 켭니다! 시스템을 종료합니다.\n")
        break
```

온라인 쇼핑몰 할인 적용하기
장바구니에 담긴 상품들의 가격 리스트가 주어져 있습니다.
다음 할인 규칙을 적용하여 각 상품의 최종 가격을 계산하고 저장하세요:
5만원 이상: 20% 할인
3만원 이상: 10% 할인
1만원 이상: 5% 할인
1만원 미만: 할인 없음
할인된 가격을 새로운 리스트에 저장하고 출력하세요.

```python
# 장바구니에 담긴 상품들의 가격
prices = [15000, 62000, 8000, 35000, 49000, 92000]
# 할인된 가격을 저장할 리스트
discounted_prices = []

# 각 상품의 할인된 가격 계산하기
for price in prices:
    # 5만원 이상 상품 할인
    if price >= 50000:
        discount = int(price * 0.2)  # 20% 할인
        final_price = price - discount
        discounted_prices.append(final_price)
        
    # 3만원 이상 상품 할인
    elif price >= 30000:
        discount = int(price * 0.1)  # 10% 할인
        final_price = price - discount
        discounted_prices.append(final_price)
        
    # 1만원 이상 상품 할인
    elif price >= 10000:
        discount = int(price * 0.05)  # 5% 할인
        final_price = price - discount
        discounted_prices.append(final_price)
        
    # 1만원 미만 상품
    else:
        discounted_prices.append(price)

# 원래 가격과 할인된 가격 출력하기
for i in range(len(prices)):
    print(f"{prices[i]}원 상품 -> {discounted_prices[i]}원")
```
