apply 메서드
[문제 1]
apply함수를 이용해 사용자 정의 함수를 시리즈에 적용해 봅시다.
그리고 그 결과를 result_series에 저장하세요.

```python
import pandas as pd

data = [1,2,3,4,5]
series = pd.Series(data)

def custom_function(x):
    if x >= 3:
        result = x+2
    else:
        result = x+4
    return result

result_series = series.apply(custom_function)
result_series
```

apply 메서드와 lambda 함수

```python
result_series = series.apply(lambda x: x+2 if x>=3 else x+4)
result_series
```

replace() 함수

```python
data = ['apple','banana','cherry']
series = pd.Series(data)

#banana -> orange
result_series = series.replace('banana','orange')
result_series
```

replace() 함수2

```python
replace_dict = {'banana':'orange','apple': 'strawberry'}
result_series = series.replace(replace_dict)
result_series
```

replace() 함수3

```python
result_series = series.replace('^a','fruit', regex=True)
result_series
#regex=True를 사용할 때 쓰이는 ^ 기호는 정규 표현식(Regular Expression)의 '메타 문자' 중 하나
#정규 표현식에서 ^는 **"문자열의 맨 처음"**을 의미
# **반드시 'a'로 시작하는 문자열"**만 찾으라는 뜻.
```
