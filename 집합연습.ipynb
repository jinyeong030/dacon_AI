집합 중복X, 순서없음(index X), 수정 가능

```python
#list 집합으로 변환
score_list = [46,85,30.90,100,55]
score_set = set(score_list)
print("중복 제거된 점수:",score_set)

text = "hello"
char_set  = set(text)
print("중복 제거된 문자:",char_set)

#빈 집합 만들기
empty_set = set()
print("빈 집합:", empty_set)

#데이터 추가
empty_set.add("1번째")
empty_set.add("2번째")

#중복 데이터 제거
student_list = ["민지","현우","민지","서연","민지"]
actual_student = set(student_list)
print("실제 수강생:",actual_student)
```

add() 요소 추가(이미 있는 값은 무시됨)
update() 여러 항목 추가
remove() 요소 삭제 - 없는 값 삭제 시도시 오류
discard() 요소 삭제 - 없는 값 삭제 시도 오류 안남
clear() 전체 요소 삭제

```python
numbers = {1,2,3}
numbers.add(4)
print("add 4 :", numbers)

numbers.update(["54","40","5"])
print("여러 항목 추가: ",numbers)

numbers.remove(3)
print("3삭제:" , numbers)

numbers.discard(3)
print("3 재삭제 시도:",numbers)

numbers.clear()
print("전체 삭제: ",numbers)
print("빈 집합인가?:", len(numbers)==0)
```

교집합 합집합 차집합
& = .intersection()
| = .union
- = .difference

```python
#두 학생이 좋아하는 과목
hee_subjects = {"국어","수학","영어","과학"}
soo_subjects = {"수학","영어","체육","미술"}

common_subjects = hee_subjects & soo_subjects
print("& 기호로 구한 공통 과목: ",common_subjects)

common_subjects2 = hee_subjects.intersection(soo_subjects)
print("intersection()으로 구한 공통 과목: ",common_subjects)

all_subjects = hee_subjects | soo_subjects
print("|기호로 구한 공통 과목: ", all_subjects)

all_subjects2 = hee_subjects.union(soo_subjects)
print("union()으로 구한 모든 과목:", all_subjects2)
```

```python
soo_movies = {"인셉션", "어벤져스", "해리포터", "매트릭스"}
hee_movies = {"인셉션", "어벤져스", "타이타닉", "노트북"}
ki_movies ={"해리포터", "타이타닉", "노트북", "인셉션"}

#3모두 본 영화
all_watched = soo_movies&hee_movies&ki_movies
print("모두 본 영화:",all_watched)

#철수와 영희가 본 영화
soo_hee_common = soo_movies & hee_movies
print("철수와 영희 둘 다 본 영화:", soo_hee_common)

#세사람이 본 모든 영화
total_movies = soo_movies |hee_movies|ki_movies
print("세사람이 본 모든 영화:", total_movies)

#soo만 본 영화
soo_only = soo_movies -(hee_movies|ki_movies)
print("soo만 본 영화:", soo_only)

```

```python
# 리스트를 집합으로 변환 (중복 제거 효과!)
num_list = [1, 2, 2, 3, 3, 3]
num_set = set(num_list)
print("중복 있는 리스트를 집합으로:", num_set)

# 튜플을 집합으로 변환 (중복 제거 효과!)
num_tuple = (1, 2, 2, 3, 3, 3)
num_set = set(num_tuple)
print("중복 있는 튜플를 집합으로:", num_set)
```
