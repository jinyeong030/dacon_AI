```python
import random
def get_random_order():
    """
    ëœë¤í•˜ê²Œ 1~3ì” ì‚¬ì´ ì£¼ë¬¸ëŸ‰ ë§Œë“¤ì–´ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜.
    """
    return random.randint(1,3)


```

```python
#ì£¼ë¬¸ì²˜ë¦¬
def handle_order(stock,order):
    if stock >= order:
        stock -= order
        print(f"{order}ì” ì£¼ë¬¸ ì™„ë£Œ! ë‚¨ì€ ì¬ê³ : {stock}ì”\n")
        return stock, True
    else:
        print(f"ì¬ê³ ê°€ ë¶€ì¡±í•©ë‹ˆë‹¤.")
        return stock ,False
```

```python
def run_coffee_shop():
    stock = 10
    print("===ì»¤í”¼ìˆ ì¬ê³  ì‹œë®¬ë ˆì´ì…˜ ì‹œì‘===")

    while True:
        order = get_random_order()
        print(f"ì´ë²ˆ ì£¼ë¬¸ ìš”ì²­ ìˆ˜ëŸ‰: {order}ì”")

        stock, success = handle_order(stock, order)
        if not success:
            print("ì£¼ë¬¸ ì¢…ë£Œ")
            break
```

```python
run_coffee_shop()
```

```python
def recommend_clothes(temperature):
    # 30ë„ ì´ìƒì¼ ë•Œ
    if temperature >= 30:
        return "ë§¤ìš° ë”ì›€, ë°˜íŒ”ê³¼ ë°˜ë°”ì§€ ì¶”ì²œ"
    
    # 20ë„ ì´ìƒ 30ë„ ë¯¸ë§Œì¼ ë•Œ
    elif temperature >= 20 and temperature < 30:
        return "ë”ì›€, ë°˜íŒ”ê³¼ ì–‡ì€ ê¸´ë°”ì§€ ì¶”ì²œ"
    
    # 10ë„ ì´ìƒ 20ë„ ë¯¸ë§Œì¼ ë•Œ
    elif temperature >= 10 and temperature < 20:
        return "ì„ ì„ í•¨, ê°€ë””ê±´ì´ë‚˜ ì–‡ì€ ê²‰ì˜· ì¶”ì²œ"
    
    # 0ë„ ì´ìƒ 10ë„ ë¯¸ë§Œì¼ ë•Œ
    elif temperature >= 0 and temperature < 10:
        return "ìŒ€ìŒ€í•¨, ìì¼“ì´ë‚˜ ì½”íŠ¸ ì¶”ì²œ"
    
    # 0ë„ ë¯¸ë§Œì¼ ë•Œ
    else:
        return "ë§¤ìš° ì¶”ì›€, ë‘êº¼ìš´ ì½”íŠ¸ì™€ ëª©ë„ë¦¬ ì¶”ì²œ"

# í…ŒìŠ¤íŠ¸
temperatures = [33.5, 28.9, 15.2, 7.1, -3.8]

for temp in temperatures:
    result = recommend_clothes(temp)
    print(f"ê¸°ì˜¨ì´ {temp}ë„ì¼ ë•Œ â†’ {result}")
```
