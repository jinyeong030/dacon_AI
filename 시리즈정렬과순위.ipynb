```python
import pandas as pd

data = [20,10,40,50,60,30,70,80]
index = ['a','b','c','d','e','f','g','h']

series = pd.Series(data, index=index, name='MySeries')
series
```

```python
#series 를 sort_values 함수를 사용하여 정렬
series_sorted = series.sort_values()
series_sorted
```

```python
#내림차순으로 정렬 ascending = 0
series_sorted = series.sort_values(ascending=0)
series_sorted
```

```python
#인덱스를 기준으로 정렬하기 sort_index()
series_sorted = series.sort_index()
series_sorted
```

```python
#인덱스를 기준으로 내림차순으로 정렬하기
series_sorted = series.sort_values(ascending = False)
series_sorted
```

```python
#내림차순으로 순위 구하기 ascending값을 False
series_rank = series.rank(ascending = False)
series_rank
```

순위 구하기 문제풀이

학생의 시험 점수를 기준으로 순위를 내림차순으로

```python
data = {'Math': 78, 'English': 92, 'Science': 85, 'History': 68}
series = pd.Series(data)

ranked_series = series.rank(ascending = False)
ranked_series
```
