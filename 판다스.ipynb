```python
import pandas as pd

#리스트로 시리즈 생성
data = [10,20,30,40,50]
series1 = pd.Series(data)

print(series1)
print(type(series1))

#배열로 시리즈 생성
import numpy as np

data2 = [10, 20, 30, 40, 50]
np_data = np.array(data)
print("\n배열의 타입:",type(np_data))

series2 = pd.Series(np_data)
print(series2)

#딕셔너리로 생성
data_dict = {
    'a':10,
    'b':20,
    'c':30,
    'd':40,
    'e':50
}
series = pd.Series(data_dict)
series
print("\n시리즈에 이름 생성")
series = pd.Series(data_dict, name= 'My_series')
series
print("\n시리즈 데이터타입 지정")
series = pd.Series(data, name='MySeries', dtype='float')
series
```

```python
#시리즈 인덱스 정의
data1 = [1, 2, 3, 4]
index = ['a', 'b', 'c', 'd']

series = pd.Series(data1, index=index, name='MySeries', dtype='int32')
series
```

시리즈의 이름 변경

```python
series.name = "new_name"
series
```

시리즈의 타입 변경 astype()함수 사용

```python
series = series.astype('float')
series
```
