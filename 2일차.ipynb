새 환자 정보 출력 시 patients_db[-1]로 접근하세요

```python
# 기본 데이터 - 병원 환자 데이터베이스
patients_db = [
    {"patient_id": "P001", "name": "김환자", "age": 65, "department": "내과", "disease": "고혈압"},
    {"patient_id": "P002", "name": "이환자", "age": 45, "department": "외과", "disease": "맹장염"},
    {"patient_id": "P003", "name": "박환자", "age": 32, "department": "소아과", "disease": "감기"}
]
```

```python
#환자 정보 조회
print(f"===첫 번째 환자 정보===")
patient_1 = patients_db[0]
print(f"환자번호: {patient_1.get('patient_id')}")
print(f"이름: {patient_1.get('name')}")
print(f"나이: {patient_1.get('age')}")
print(f"진료과: {patient_1.get('department')}")
print(f"질병: {patient_1.get('disease')}")
```

```python
#환자 정보 수정
patient_2 = patients_db[1]
print("===질병 정보 수정 전===")
print(f"환자번호: {patient_2.get('patient_id')}")
print(f"이름: {patient_2.get('name')}")
print(f"기존 질병: {patient_2.get('disease')}")

patient_2.update({'disease':"급성맹장염"})
print(f"\n===질병 정보 수정 후===")
print(f"환자번호: {patient_2.get('patient_id')}")
print(f"이름: {patient_2.get('name')}")
print(f"기존 질병: {patient_2.get('disease')}")
```

```python
#새 환자 등록 새 환자 정보 출력 시 patients_db[-1]로 접근
new_patient = {"patient_id":"P004","name":"홍환자","age":40,"department":"피부과","disease":"아토피"}
patients_db.append(new_patient)
latest_patient = patients_db[-1]
print(f"\n===새 환자 등록===")
print(f"환자번호: {latest_patient.get('patient_id')}")
print(f"이름: {latest_patient.get('name')}")
print(f"나이: {latest_patient.get('age')}")
print(f"진료과: {latest_patient.get('department')}")
print(f"질병: {latest_patient.get('disease')}")

print(f"현재 총 환자 수: {len(patients_db)}명")
```

```python
#환자 퇴원 처리 pop()

del_patient = patients_db.pop()

print("===환자 퇴원 처리===")
print(f"퇴원 환자번호: {del_patient.get('patient_id')}")
print(f"퇴원 환자명: {del_patient.get('name')}")
print(f"퇴원 후 총 환자 수 {len(patients_db)}")
```

```python
#통계 정보 (max, min, sum)
from operator import itemgetter
print("===환자 나이 통계===")
ages = list(map(itemgetter("age"), patients_db))

print(f"최고령 환자 나이: {max(ages)}세")
print(f"최연소 환자 나이: {min(ages)}세")
print(f"평균 나이: {sum(ages)}세")

print("\n===============================")

print("최종 병원 환자 데이터베이스 현황")
print("\n===============================")
print(f"총 입원 환자: {len(patients_db)}")
```
